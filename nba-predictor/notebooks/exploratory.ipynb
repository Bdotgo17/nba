# NBA Predictor Exploratory Data Analysis

This notebook is used for exploratory data analysis (EDA) on NBA data. The goal is to visualize data, understand patterns, and test hypotheses that will inform the predictive modeling process.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')
```

```python
# Load the data
data_path = '../data/processed/nba_data.csv'  # Adjust the path as necessary
nba_data = pd.read_csv(data_path)

# Display the first few rows of the dataset
nba_data.head()
```

```python
# Summary of the dataset
nba_data.info()
nba_data.describe()
```

```python
# Visualizations
# Example: Distribution of points scored
plt.figure(figsize=(10, 6))
sns.histplot(nba_data['points'], bins=30, kde=True)
plt.title('Distribution of Points Scored')
plt.xlabel('Points')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = nba_data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
